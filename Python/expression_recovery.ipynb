```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import matplotlib.gridspec as gridspec

# Setup
methods = ['scImpute', 'DrImpute', 'scRMD', 'ALRA', 'MAGIC', 'SAVER', 'DeepImpute', 'scDoc', 'scDDI']
datasets = [
    'Masked Darmanis (5%)',
    'Masked Goolam (5%)',
    'Masked Pollen (5%)',
    'Masked Yan (5%)',
    'Masked Ting (5%)',
    'Masked Melanoma (5%)'
]

# Sample metric data (replace with your actual values)
rmse = A matrix containing RMSE values

mae = A matrix containing MAE values
# Prepare DataFrame
records = []
for i, method in enumerate(methods):
    for j, dataset in enumerate(datasets):
        idx = j * len(methods) + i
        records.append({
            'Method': method,
            'Dataset': dataset,
            'RMSE': rmse[idx],
            'MAE': mae[idx]
        })

df = pd.DataFrame(records)

# Plot settings
sns.set(style="whitegrid", palette="tab10")

# ---------- RMSE Bar Plot PDF ----------
fig = plt.figure(figsize=(18, 10))
gs = gridspec.GridSpec(2, 3)

for i, dataset in enumerate(datasets):
    ax = fig.add_subplot(gs[i // 3, i % 3])
    sns.barplot(data=df[df['Dataset'] == dataset], x='Method', y='RMSE', ax=ax)
    ax.set_title(dataset, fontsize=14)
    ax.set_xticklabels(ax.get_xticklabels(), rotation=45)

plt.suptitle("RMSE Across Imputation Methods", fontsize=18)
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()

# ---------- MAE Bar Plot PDF ----------
fig = plt.figure(figsize=(18, 10))
gs = gridspec.GridSpec(2, 3)

for i, dataset in enumerate(datasets):
    ax = fig.add_subplot(gs[i // 3, i % 3])
    sns.barplot(data=df[df['Dataset'] == dataset], x='Method', y='MAE', ax=ax)
    ax.set_title(dataset, fontsize=14)
    ax.set_xticklabels(ax.get_xticklabels(), rotation=45)

plt.suptitle("MAE Across Imputation Methods", fontsize=18)
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```
