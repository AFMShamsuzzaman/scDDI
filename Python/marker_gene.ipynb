```python
import numpy as np
import pandas as pd
import scanpy as sc
adata=sc.read_csv('darmanis_imp_t.csv', delimiter=',', first_column_names=None, dtype='float32')
```

```python
adata
```

```python
sc.pp.filter_cells(adata, min_genes=200)
sc.pp.filter_genes(adata, min_cells=3)

sc.pp.normalize_total(adata, target_sum=1e4)

sc.pp.log1p(adata)

sc.pp.scale(adata, max_value=10)
```

```python
sc.tl.pca(adata, svd_solver='arpack')
sc.pl.pca_variance_ratio(adata,n_pcs=20,log=True)
```

```python
#create neighborhood graph using 20 pcs 
sc.pp.neighbors(adata, n_neighbors=15, n_pcs=30)
##dim reduction using umap
sc.tl.umap(adata)
#Leiden clustering
import leidenalg
sc.tl.leiden(adata)
##visualizing clusters
sc.pl.umap(adata, color=['leiden'])
```

```python
pd.DataFrame(adata.obs).to_csv("darmanis_leiden.csv")
```

```python
sc.tl.rank_genes_groups(adata, 'leiden', method='wilcoxon',key_added = "wilcoxon")
sc.pl.rank_genes_groups(adata, n_genes=30, sharey=False,key="wilcoxon")
```

```python
#save top 10 DE genes for each cluster using wilcox-ranksum test
result = adata.uns['wilcoxon']
groups = result['names'].dtype.names
p=pd.DataFrame(
    {group + '_' + key[:1]: result[key][group]
    for group in groups for key in ['names', 'pvals']}).head(10)
pd.DataFrame(p).to_csv("darmanis_marker.csv")
```
